### Dataset Description 
- EEG_Data_Pivoted.csv: Represents the EEG signals of subjects from the alcoholic group.
- EEG_Data_Control_Pivoted.csv: Represents the EEG signals of subjects from the control group.
- Each file is pivoted with the time axis (x-axis) and sensor values from 61 electrodes (y-axis).

### Column Descriptors
1. time:
- Type: Numeric
- Description: Time in seconds. It represents the x-axis in the data, sampled at 256 Hz, spanning 1 second (256 samples).

2. sensor_<x> (e.g., sensor_1, sensor_2, …, sensor_61):
- Type: Numeric
- Description: Sensor values in microvolts recorded from 61 electrodes placed on the scalp. Each column corresponds to a specific electrode sensor.

```python
alcoholic_df = pd.read_csv('data/EEG_Alcohol_Data/EEG_data_pivoted.csv')
control_df = pd.read_csv('data/EEG_Alcohol_Data/EEG_data_control_pivoted.csv')
```

### Alcoholism VS Control
- Label = 1 (Alcoholism)
- Label = 0 (Control)

```python
alcoholic_df['label'] = 1
control_df['label'] = 0
```

```python
combined_df = pd.concat([alcoholic_df , control_df] , ignore_index=True)
combined_df.head()
```
